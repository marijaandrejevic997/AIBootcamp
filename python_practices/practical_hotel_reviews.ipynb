```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import pandas as pd
import re

```

```python
data = pd.read_csv('tripadvisor_hotel_reviews.csv')
data.head()
```

```python
#data preprocessing
#lowercase everything and add column with lowercase reviews
data['lower_case_Review'] = data['Review'].str.lower()
data.head()
```

```python
# stopwords - common words that do not add much meaning to a sentence, such as "the", "is", "in", etc. We can remove them to focus on the more meaningful words in the reviews.
stop_words = set(stopwords.words('english'))
stop_words.remove('not') # we want to keep "not" because it can change the meaning of a sentence, for example "not good" is different from "good"
data['no_stopwords_Review'] = data['lower_case_Review'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
data.head()
```

```python
data['Review'][0]
```

```python
data['lower_case_Review'][0]
```

```python
data['no_stopwords_Review'][0]
```

```python
#replace * with star because * can be used to indicate a rating, for example "5* hotel" means a 5 star hotel, and we want to keep that information in the review
data['cleaned_Review'] = data.apply(lambda x: re.sub(r'[*]', ' star', x['no_stopwords_Review']), axis=1)
data.head()
```

```python
data['cleaned_Review'][0]
```

```python
data['cleaned_Review'] = data['cleaned_Review'].apply(lambda x: re.sub(r'[^\w\s]', '', x)) # remove punctuation isn't a work or space
```

```python
data.head()
```

```python
data['tokenized'] = data['cleaned_Review'].apply(word_tokenize)
data.head()
```

```python
data['tokenized'][0]
```

```python
ps = PorterStemmer()
data['stemmed'] = data['tokenized'].apply(lambda x: [ps.stem(word) for word in x])
data.head()
```

```python
data['stemmed'][0]
```

```python
lemmatizer = WordNetLemmatizer()
data['lemmatized'] = data['tokenized'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])
data.head()
```

```python
data['lemmatized'][0]
```

```python
tokens_cleaned = sum(data['lemmatized'], [])
```

```python
unigrams = (pd.Series(nltk.ngrams(tokens_cleaned, 1)).value_counts())
print(unigrams)
```

```python
bigrams = pd.Series(nltk.ngrams(tokens_cleaned, 2)).value_counts()
print(bigrams)
```
